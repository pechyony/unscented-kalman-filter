This notebook shows graphs of normalized innovation scores of radar and laser sensors and compares them with 95% percentile. These graphs were used to tune standard deviations of process noise. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
laser_nis = np.loadtxt('laser_nis_file.txt')
radar_nis = np.loadtxt('radar_nis_file.txt')
```

```python
plt.figure(figsize=(17,7))
# we exclude first innovation score of the radar since it is computed after getting the second measurement and 
# is based on an estimate of velocity from a single laser measurement  
plt.plot(radar_nis[1:])
plt.plot((0, 249), (7.815, 7.815), 'r-')
plt.ylabel('normalized innovation score')
plt.xlabel('time step')
plt.title('Radar measurements')
plt.text(238, 8.2, "95% percentile", color="r")
plt.show()
```

In the above graph 14 values are above 95% percentile. 14 / 250 * 100 = 5.6%. Hence the filter is calibrated w.r.t. the radar measurements. 

```python
plt.figure(figsize=(17,7))
# first innovation score of laser is computed after getting the third measurement 
plt.plot(laser_nis)
plt.plot((0, 249), (5.991, 5.991), 'r-')
plt.ylabel('normalized innovation score')
plt.xlabel('time step')
plt.title('Laser measurements')
plt.text(238, 6.2, "95% percentile", color="r")
plt.show()
```

In the above graph 10 values are above or very close to 95% percentile. 10 / 250 * 100 = 4%. Hence the filter is calibrated w.r.t. the laser measurements.
